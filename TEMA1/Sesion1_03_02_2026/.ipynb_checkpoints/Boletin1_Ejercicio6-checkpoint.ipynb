```python
import numpy as np

#Definimps la funcion
def f(p):
    x=p[0]
    y=p[1]
    val=np.log(x-2*y)
    return val

print('f(3,1) =',f([3,1]))

#Funcion que calcula el gradiente en cada pto p
def g_ex(p):
    x=p[0]
    y=p[1]
    grad=np.zeros_like(p)
    grad[0]=1/(x-2*y); grad[1]=-2/(x-2*y)
    return grad

print('grad(3,1) =',g_ex([3,1]))

#Polinomio de taylor centrado en (3,1), a=centro y b=pto donde se evalua
def P(a,p):
    val=f(a)+np.dot(g_ex(a),p-a) #np.dot producto escalar = producto punto
    return val
a=np.array([3,1],dtype=float)
p=np.array([3.1,1.05],dtype=float)

print('Valor exacto f(3.1,1.05)',f(p))
print('Valor aproximado P(3.1,1.05) =',P(a,p))
```
