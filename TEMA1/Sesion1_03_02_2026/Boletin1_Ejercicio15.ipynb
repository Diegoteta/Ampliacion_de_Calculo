```python
import numpy as np

#Definimps la funcion
def f(p):
    x=p[0]
    y=p[1]
    val=np.exp(x)*np.sin(y)
    return val

print('f(0,0) =',f([0,0]))

#Funcion que calcula el gradiente en cada pto p
def g_ex(p):
    x=p[0]
    y=p[1]
    grad=np.zeros_like(p)
    grad[0]=np.exp(x)*np.sin(y); grad[1]=np.exp(x)*np.cos(y)
    return grad

print('grad(0,0) =',g_ex([0,0]))

#Polinomio de taylor centrado en (0,0),McLaurin, a=centro y b=pto donde se evalua
def P(a,p):
    val=f(a)+ g_ex(a) @ (p-a) #np.dot producto escalar = producto punto
    return val
a=np.array([0,0],dtype=float)
p=np.array([1/10,1/20],dtype=float)

print('Valor exacto f(1/10,1/20)',f(p))
print('Valor aproximado P(3.1,1.05) =',P(a,p))
```
