```python
import numpy as np

#Definimps la funcion
def f(p):
    x=p[0]
    y=p[1]
    val=np.log(1+x**2+y**2)+np.exp(x*y)
    return val
    
#Derivada parcial rexpecto de x de f en el pto p
def dx_ex(p):
    x=p[0]
    y=p[1]
    val=(2*x)/(1+x**2+y**2)+y*np.exp(x*y)
    return val
print('Derivada exacta en x (1,1) =',dx_ex([1.,-1.]))

#Aproximacion numerica de la parcial respecto de x
def dx_num(p):
    h=1e-6
    val=f([p[0]+h,p[1]])-f([p[0],p[1]]); val=val/h
    return val
print('Derivada aprox_num en x (1,1) =',dx_num([1.,-1.]))
print('Error en dx =',np.abs(dx_ex([1,1])-dx_num([1,1])))  
#Derivada parcial rexpecto de y de f en el pto p
def dy_ex(p):
    x=p[0]
    y=p[1]
    val=(2*y)/(1+x**2+y**2)+x*np.exp(x*y)
    return val
print('Derivada exacta en y (1,1) =',dy_ex([1.,-1.]))

#Aproximacion numerica de la parcial respecto de y
def dy_num(p):
    h=1e-6
    val=f([p[0],p[1]+h])-f([p[0],p[1]]); val=val/h
    return val
print('Derivada aprox_num en y (1,1) =',dy_num([1.,-1.]))
print('Error en dy =',np.abs(dy_ex([1,1])-dy_num([1,1])))

#Gradiente exacto
def grad_ex(p):  
    grad=np.zeros_like(p)  
    grad[0]=dx_ex(p)  
    grad[1]=dy_ex(p)  
    return grad
print('Gradiente exacto en (1,1)',grad_ex([1.,-1.]))

#Gradiente aproximado_numerico
def grad_num(p): 
    grad=np.zeros_like(p)  
    grad[0]=dx_num(p)  
    grad[1]=dy_num(p) 
    return grad
print('Gradiente aprox_num en (1,1)',grad_num([1.,-1.]))
print('Error gradiente =',np.abs(grad_ex([1,-1])-grad_num([1,-1])))

#Derivada direccional exacta
def Du_ex(p,v):  
    u=v/np.linalg.norm(v) 
    grad=grad_ex(p) 
    val=grad @ u  
    return val
Ddir_ex=Du_ex(np.array([1.,1.]),np.array([2.,1.]))
print('Derivada direccional exacta =',Ddir_ex)

#Derivada direccional aproximada_num
def Du_num(p,v):  
    h=1e-6  
    u=v/np.linalg.norm(v)  
    u1=u[0]  
    u2=u[1] 
    val=(f(p+h*u)-f(p))/h  
    return val
Ddir_num=Du_num(np.array([1.,1.]),np.array([2.,1.]))
print('Derivada direccional aproximada_num =',Ddir_num)
```
